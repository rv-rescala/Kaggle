```python
import pandas as pd

# サンプルデータフレームを作成
df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar'],
                   'B': [1, 2, 3, 4]})

# グループ化してtransformを適用
df['transformed'] = df.groupby('A')['B'].transform('mean')

d = df.groupby(by=['A']).apply(lambda x: x['B'].mean()).to_dict()
df['applied'] = df.apply(lambda x: d[x['A']], axis=1)

df
```
